```python
import re
```

```python
df1['new']= df1['molecule_name'].apply(lambda x: (re.sub('[A-Z]','',x)))
```

```python
df1['new1']= df1['new'].apply(lambda x: (re.sub('-','',x)))
```

```python
df1.drop(["new","new1","molecule_name"],axis=1,inplace=True)
```

```python
df1['new2']=df1['new1'].apply(lambda x: (re.sub('[a-z]','',x)))
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df1.drop(["class"],axis=1),df1['class'],random_state=102,test_size=0.2)
```

```python
x_train.info()
```

```python
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(x_train, y_train)
```

```python
pred_lg=lg.predict(x_test)
from sklearn.metrics import classification_report
from sklearn.metrics import roc_auc_score
print(roc_auc_score(y_test,pred_lg))
print(classification_report(y_test,pred_lg))
lg.score(x_train, y_train)
```

```python
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,pred_lg)
cm
```

```python
from sklearn.ensemble import RandomForestClassifier
rf=RandomForestClassifier()
rf.fit(x_train, y_train)
rf_pred=rf.predict(x_test)
print(classification_report(y_test,rf_pred))
rf.score(x_train, y_train)
```
